# Scraping Cognitive Atlas

This notebook scrapes concept terms from Cognitive Atlas (https://www.cognitiveatlas.org/concepts), and saves the file in ../data/cogatlas_terms.txt. 

This is the first thing to be done, only has to be done once, and these terms are used as seeds for the rest of the analyses.

In theory, could be extended to the other categories on Cognitive Atlas, though the concepts are the most filled out.

```python
import numpy as np
from bs4 import BeautifulSoup
import urllib

cog_atl_url = 'http://www.cognitiveatlas.org/concepts'

# scraping the cognitive atlas webpage for terms
CAurl = urllib.request.urlopen(cog_atl_url).read()
soup = BeautifulSoup(CAurl, 'html.parser')
concepts = soup.find_all('span', attrs={'class': 'concept'})
terms = []
for c in concepts:
    terms.append(str(c.text.strip()).lower())
    
terms.append('') # append empty term to calculate total article count
print(len(terms))
```

```python
terms
```

```python
# save terms to disk
term_file = '../data/cogatlas_terms.txt'
file = open(term_file, 'w')
for t in terms:
    file.write("%s\n" % t)
    
file.close()
```
